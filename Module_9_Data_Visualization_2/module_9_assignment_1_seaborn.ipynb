```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Do the following tasks:** 
* set style to '`white`'
* load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style("white")
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic.head()
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
tips.head()
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4, 3))
sns.boxplot(x='day', y='tip', hue='sex', data=tips, palette={'Male': 'skyblue', 'Female': 'green'})
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', height=5)
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
titanic['fare'].hist(bins=20,color='green')
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(6, 3))
sns.boxplot(x='class', y='age', hue='alive', data=titanic, palette={'yes': 'orange', 'no': 'skyblue'})
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(x='fare', y='age', data=titanic, kind='scatter', height=5) 
```

**Recreate the plot below. Use titanic dataset. ** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4,3))
palette = {'First': 'skyblue', 'Second': 'lightgreen', 'Third': 'orange'}
sns.boxplot(x='class', y='age', hue='class', data=titanic, palette=palette)
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4, 3))

sns.countplot(x='sex', hue='class', data=titanic,palette={'First': 'blue', 'Second': 'green', 'Third': 'darkred'})
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(6,4))
data = titanic[['survived','pclass','age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']].corr()
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('titanic.corr()')
plt.show()
```
