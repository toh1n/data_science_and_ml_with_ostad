### Read data from income.csv and scale the relevant features using standardization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

sns.histplot(df['Age'], kde=True, kde_kws=dict(cut=3), ax=axes[0])
axes[0].set_title('Age Distribution')

sns.histplot(df['Income'], kde=True, kde_kws=dict(cut=3), ax=axes[1])
axes[1].set_title('Income Distribution')

plt.tight_layout()
plt.show()
```

# Without using library

```python
df1 = df.copy()
df1['Age'] = (df1['Age'] - df1['Age'].mean()) / df1['Age'].std()
df1['Income'] = (df1['Income'] - df1['Income'].mean()) / df1['Income'].std()
df1.head()
```

```python
# sns.distplot(df1['Age'])
# sns.histplot(df1['Age'],kde=True, kde_kws=dict(cut=3))
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

sns.histplot(df1['Age'], kde=True, kde_kws=dict(cut=3), ax=axes[0])
axes[0].set_title('Age Distribution')

sns.histplot(df1['Income'], kde=True, kde_kws=dict(cut=3), ax=axes[1])
axes[1].set_title('Income Distribution')

plt.tight_layout()
plt.show()
```

# Using library

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
df2 = df.copy()
df2[['Age', 'Income']] = std.fit_transform(df[['Age', 'Income']])
df2.head()

```

```python
# sns.distplot(df2['Age'])
# sns.histplot(df2['Age'],kde=True, kde_kws=dict(cut=3))

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

sns.histplot(df2['Age'], kde=True, kde_kws=dict(cut=3), ax=axes[0])
axes[0].set_title('Age Distribution')

sns.histplot(df2['Income'], kde=True, kde_kws=dict(cut=3), ax=axes[1])
axes[1].set_title('Income Distribution')

plt.tight_layout()
plt.show()
```
